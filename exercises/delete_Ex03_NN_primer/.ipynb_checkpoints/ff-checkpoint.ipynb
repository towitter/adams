```python
################## DNN CV ##################

### Select data
data=pd.read_csv("newset.csv",engine='python', encoding='utf-8', error_bad_lines=False)
data=data.drop(['Unnamed: 0'], axis=1)
data=data.dropna()
#X=data.drop(['Target100',"TotalPL_GBP100",'AvgRetNext100'], axis=1).astype(float)
X=data.drop(['Target100','AvgRetNext100'], axis=1).astype(float)
money=X["TotalPL_GBP20"]
#X=data.drop(['Target100'], axis=1).astype(float)
X=X.values
y=np.ravel(data['Target100']) 
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
seed=5
### Define model

def create_network():
    
    # Start neural network
    network = models.Sequential()

    # Add fully connected layer with a ReLU activation function
    network.add(Dense(1024, activation = 'relu', input_shape=(X.shape[1],)))
    network.add(BatchNormalization())
    network.add(Dropout(0.5))
    network.add(Dense(units = 1024, kernel_initializer=keras.initializers.glorot_normal(seed=seed), activation = 'relu'))
    network.add(BatchNormalization())
    network.add(Dropout(0.5))
    network.add(Dense(units =1024, kernel_initializer=keras.initializers.glorot_normal(seed=seed), activation = 'relu'))
    network.add(BatchNormalization())
    network.add(Dropout(0.5))
    network.add(Dense(units = 1, kernel_initializer=keras.initializers.glorot_normal(seed=seed), activation = 'sigmoid'))


    # Compile neural network
    network.compile(loss='binary_crossentropy', # Cross-entropy
                    optimizer='adam', # Root Mean Square Propagation
                    metrics=['accuracy']) # Accuracy performance metric
    
    # Return compiled network
    return network

### Prepare CV

cv = StratifiedKFold(n_splits=10,random_state=5)
### Select classifier

classifier = KerasClassifier(build_fn=create_network, 
                                 epochs=10, 
                                 batch_size=512, 
                                 verbose=1)

### Set lists for records

tprs = []
aucs_dnn = []
profit_dnn=[]
f1_dnn=[]
mean_fpr = np.linspace(0, 1, 100)

### Start the CV
i = 0
for train_index, test_index in cv.split(X,y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    #smt = SMOTE(sampling_strategy='auto',random_state=5,k_neighbors=5)
    #X_train, y_train = smt.fit_sample(X_train, y_train)
    pnl=money[test_index].values
    scaler = StandardScaler().fit(X_train)
    X_train= scaler.transform(X_train)
    X_test= scaler.transform(X_test)  
    classifier.fit(X_train, y_train)
    y_pred = np.around(classifier.predict(X_test) , decimals=0)
    yhat=y_pred.flatten()
    gain=np.sum(pnl)-np.sum(np.multiply(yhat, pnl))
    profit_dnn.append(gain)
    fpr, tpr, thresholds = roc_curve(y_test,  y_pred)
    tprs.append(interp(mean_fpr, fpr, tpr))
    tprs[-1][0] = 0.0
    roc_auc = auc(fpr, tpr)
    aucs_dnn.append(roc_auc)
    f1_dnn.append(classification_report(y_test, y_pred,output_dict=True)["1"]["f1-score"])  
    plt.plot(fpr, tpr, lw=1, alpha=0.3,
             label='ROC fold %d (AUC = %0.2f)' % (i, roc_auc))



    i += 1
plt.plot([0, 1], [0, 1], linestyle='--', lw=2, color='r',
         label='Chance', alpha=.8)

mean_tpr = np.mean(tprs, axis=0)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
std_auc = np.std(aucs_dnn)
plt.plot(mean_fpr, mean_tpr, color='b',
         label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_auc, std_auc),
         lw=2, alpha=.8)

std_tpr = np.std(tprs, axis=0)
tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
plt.fill_between(mean_fpr, tprs_lower, tprs_upper, color='grey', alpha=.2,
                 label=r'$\pm$ 1 std. dev.')

plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```
