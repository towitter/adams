## Keras 'flow' and data augmentation: ImageDataGenerator

![struct](https://miro.medium.com/max/700/1*HpvpA9pBJXKxaPCl5tKnLg.jpeg)

Keras offers a relatively convenient wrapper, that can only be used given a certain data structure. It has an additional helpful trick:  **image augmentation**. It allows to expand the data set and populate it with 'transformed' (rotation, sheering, etc) photos of the represented labels and thus, help teach model to generalize. 
![imageaug](https://miro.medium.com/max/2500/1*rvwzKkvhlDN3Wo_4Oay_4Q.png)

Danger! Danger! By now you should have noticed that something is going terribly wrong... Doing the image augmentation on the full dataset will unnecessarily reduce the performance on the test data! We wouldn't flip and rotate the images that we want to predict after deployment. Instead we should use two separate generators, as image augmentation should **only** be done on a training data! Mind that you will need a certain folder structure for it (https://keras.io/preprocessing/image/)

```python
#Let's be thoughtful about parameters
sqrt_dim=50
train_batch_size=2
input_shape=(sqrt_dim, sqrt_dim, 3)
```

```python
train_generator = ImageDataGenerator(rescale = 1./255, 
                                   shear_range = 0.1, 
                                   zoom_range = 0.1,
                                   horizontal_flip = True) 

test_generator = ImageDataGenerator(rescale = 1./255) # only rescaling with fixed parameters is valid, otherwise do not touch the test set! 

trainset = train_generator.flow_from_directory('faces/training', # specify the directory (you can right click on the folder you need and copy the path)
                                                 target_size = (sqrt_dim, sqrt_dim), # you need to check your data,
                                                 batch_size = train_batch_size, # batch size = 5 here just for show!
                                                 class_mode= 'categorical')

# You will have to create your own validation set(s) when working with data from a third party
# validationset = [...]

testset = test_generator.flow_from_directory('faces/testing',  # specify the directory
                                            target_size = (sqrt_dim,sqrt_dim),
                                            batch_size = 1, #to sample an image just once
                                            class_mode = 'categorical')
```

```python
n_train_samples = len(trainset.filenames)
n_classes = len(trainset.class_indices)
print(n_train_samples, n_classes)
```

## CNN refresher

Playground to look into the CNN architecture: http://scs.ryerson.ca/~aharley/vis/conv/flat.html

```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten, Dropout # converting pooled feature maps into a large feature vector to prepare for input into the Dense layer
from keras.layers import Conv2D # first convolution step, 3D for videos
from keras.layers import MaxPooling2D #pooling, downsampling

model = Sequential() #initialize
model.add(Conv2D(32, #number of filters/feature detectors, number of feature maps you will end up with. Starting with 32 is good practice, grow it if needed and if you have enough computing power
                 kernel_size=(5, 5), # size of filters (rows,columns), (3,3) is also a popular choice, it's a hyperparameter to choose
                 activation='relu', # removes the negative pixels, introducing non-linearity, 
                 #we are skipping "border mode" - it is set to "same" by default
                 strides=(2, 2),# how the window shifts by in each of the dimensions. Default stride is 1 but we will go for 2 for faster implementation by reducing the image size similar to pooling
                 input_shape=input_shape)) # (3 if color/1 if b&w, height, width), all images have to come in a single format
model.add(MaxPooling2D(pool_size=(2, 2), #the size of the pooled feature map
                       strides=None)) # default: pool_size, here: (2,2)
model.add(Conv2D(64, (5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(16, (1, 1), activation='relu'))
model.add(Flatten())
model.add(Dense(512, activation='relu')) # that's our classifier bit, th size is up to you, choose wisely taking the size of picture in mind (smth between nuber of input and number of output nodes could be another rule of thumb)
model.add(Dropout(0.2))
model.add(Dense(n_classes, activation='softmax')) 
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

Little trick: train until certain accuracy (though be careful what you "wish for")

```python
import tensorflow as tf
class AccuracyCallback(tf.keras.callbacks.Callback):
    """Callback that stops training on requested accuracy."""
    def __init__(self, target_accuracy=0.997):
        self.target_accuracy = target_accuracy
    def on_epoch_end(self, epoch, logs):
        if logs.get('val_acc') >= self.target_accuracy:
            print("Target validation accuracy (%f) reached !" % self.target_accuracy )
            self.model.stop_training = True
            
target_accuracy = 0.7
max_epochs = 40
print("\nTraining untill accuracy reaches %s or for %i epochs" %(target_accuracy, max_epochs))
```

We have a special wrapper supplied by Keras

```python
history=model.fit_generator(trainset,
                    steps_per_epoch= 190,#how many batches of samples will generator provide before finishing the epoch, normally numOfObs/bacth_size
                    validation_data=testset,
                    validation_steps= 170, # 340/2
                    epochs=max_epochs, 
                    callbacks=[AccuracyCallback(target_accuracy)]
)
#steps_per_epoch is used to generate the entire dataset once by calling the generator steps_per_epoch times, the smaller the faster
#generator creates one augmented by one normal
```

```python
def show_history(history):
    plt.plot(history.history['acc'])
    plt.plot(history.history['val_acc'])
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train_accuracy', 'test_accuracy'], loc='best')
    plt.show()

show_history(history)
```

```python
model.evaluate_generator(generator=testset,steps=140)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

As always, checking for dimensions

```python
print(X_train.shape)
print(y_train.shape)
```

We will need an extra dimension (1400, 50, 37, 1) for the proper functionning of the CNN

```python
#make sure you get the format right before you feed it in, 128x128 size could be very nice but computationally hard
X_train = X_train.reshape(X_train.shape[0], lfw_people.images.shape[1], lfw_people.images.shape[2],1)
X_test = X_test.reshape(X_test.shape[0], lfw_people.images.shape[1], lfw_people.images.shape[2],1)
```

Order of input shape parameters can vary, depending on theano or tf backend, we use tf so we finish with number of channels!

```python
input_shape = (lfw_people.images.shape[1], lfw_people.images.shape[2],1) # we would have 3 instead of one if images were colored
```

```python
X_train = X_train.astype('float64')
X_train /= 255

X_test = X_test.astype('float64')
X_test /= 255
```

```python
from keras.utils import np_utils
y_train = np_utils.to_categorical(y_train, 19)
y_test = np_utils.to_categorical(y_test, 19)
```

Our model remains the same

```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten # converting pooled feature maps into a large feature vector to prepare for input into the Dense layer
from keras.layers import Conv2D # first convolution step, 3D for videos
from keras.layers import MaxPooling2D #pooling, downsampling

model2 = Sequential() #initialize
model2.add(Conv2D(32, #number of filters/feature detectors, number of feature maps you will end up with. Starting with 32 is good practice, grow it if needed and if you have enough computing power
                 kernel_size=(5, 5), # size of filters (rows,columns), (3,3) is also a popular choice, it's a hyperparameter to choose
                 activation='relu', # removes the negative pixels, introducing non-linearity, 
                 #we are skipping "border mode" - it is set to "same" by default
                 strides=(2, 2),# how the window shifts by in each of the dimensions. Default stride is 1 but we will go for 2 for faster implementation by reducing the image size similar to pooling
                 input_shape=input_shape)) # (3 if color/1 if b&w, height, width), all images have to come in a single format
model2.add(MaxPooling2D(pool_size=(2, 2), #the size of the pooled feature map
                       strides=(2,2))) # Nonecould also be an option, default: pool_size, here: (2,2)
model2.add(Conv2D(64, (5, 5), activation='relu'))
model2.add(MaxPooling2D(pool_size=(2, 2)))
model2.add(Conv2D(16, (1, 1), activation='relu'))
model2.add(Flatten())
model2.add(Dense(512, activation='relu')) # that's our classifier bit, th size is up to you, choose wisely taking the size of picture in mind (smth between nuber of input and number of output nodes could be another rule of thumb)
model2.add(Dense(n_classes, activation='softmax')) 
```

```python
model2.summary()
```

```python
batch_size=64
epochs=50

model2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
history=model2.fit(X_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_split=0.2)
```

```python
def show_history(history):
    plt.plot(history.history['acc'])
    plt.plot(history.history['val_acc'])
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train_accuracy', 'test_accuracy'], loc='best')
    plt.show()

show_history(history)
```

```python
score = model2.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
from keras.models import model_from_json
from keras.models import Model
from keras.layers import Input, Convolution2D, ZeroPadding2D, MaxPooling2D, Flatten, Dense, Dropout, Activation

model = Sequential()
model.add(ZeroPadding2D((1,1),input_shape=(224,224, 3)))
model.add(Convolution2D(64, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(128, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(256, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(256, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(256, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(Convolution2D(4096, (7, 7), activation='relu'))
model.add(Dropout(0.5))
model.add(Convolution2D(4096, (1, 1), activation='relu'))
model.add(Dropout(0.5))
model.add(Convolution2D(2622, (1, 1)))
model.add(Flatten())
model.add(Activation('softmax'))
```

Now let's load the weights. Previous layer of the output layer is the one for representation. Our next actions will for output of that layer.

Some helper functions:

```python
def findCosineSimilarity(source_representation, test_representation):
    a = np.matmul(np.transpose(source_representation), test_representation)
    b = np.sum(np.multiply(source_representation, source_representation))
    c = np.sum(np.multiply(test_representation, test_representation))
    return 1 - (a / (np.sqrt(b) * np.sqrt(c)))
```

```python
epsilon = 0.6 # let's set a threshold for cosine similarity, the the smaller the distance between two vectors, the more similair they are. Bigger epsilon will be more forgiving.

def verifyFace(img1, img2):
    img1_representation = vgg_face_descriptor.predict(preprocess_image('faces/%s' % (img1)))[0,:]
    img2_representation = vgg_face_descriptor.predict(preprocess_image('faces/%s' % (img2)))[0,:]
    
    cosine_similarity = findCosineSimilarity(img1_representation, img2_representation)
    
    print("Cosine similarity: ",cosine_similarity)
    
    if(cosine_similarity < epsilon):
        print("verified... they are same person")
    else:
        print("unverified! they are not same person!")
    
    f = plt.figure()
    f.add_subplot(1,2, 1)
    plt.imshow(image.load_img('faces/%s' % (img1)))
    plt.xticks([]); plt.yticks([])
    f.add_subplot(1,2, 2)
    plt.imshow(image.load_img('faces/%s' % (img2)))
    plt.xticks([]); plt.yticks([])
    plt.show(block=True)
    print("-----------------------------------------")
```
